```python
graph={}
ed=int(input("enter the number of edges: "))
for i in range(ed):
  x,wei,y=input().split()
  val=(wei,y)
  if x in graph:
    graph[x].append(val)
  else:
    graph[x]=[]
    graph[x].append(val)
print(graph)
```

```python
from queue import PriorityQueue
```

```python

st_node=input("enter the first node: ")
en_node=input("enter the goal node: ")
visited=[]
par={}
fin=[]
def ucs(st_node, en_node):
  que=PriorityQueue()
  que.put((0,st_node))
  while not que.empty():
    cur_pair=que.get()
    cur_node=cur_pair[1]
    print(cur_pair)
    if cur_node==en_node:
      que.queue.clear()
    else:
      if cur_node in graph:
        if cur_node not in visited:
          visited.append(cur_node)
          for nei in graph[cur_node]:
            x=int(cur_pair[0])+int(nei[0])
            que.put((x,nei[1]))
            par[nei[1]]=cur_node

  #print(cur_node)
  fin.append(cur_node)
  #print(par)
  while cur_node in par:
      cur_node=par[cur_node]
      #print(cur_node)
      fin.append(cur_node)
  while fin:
    x=fin.pop()
    print(x)

ucs(st_node,en_node)
```
