```python
from queue import PriorityQueue
import itertools # Import itertools

# -------------------- Initial & Goal --------------------

start =start = [
 [1,4,7],
 [2,5,8],
 [3,6,0]
]


goal = [
 [1,2,3],
 [4,5,6],
 [8,7,0]
]



# -------------------- Utilities --------------------

def to_tuple(state):
    return tuple(tuple(row) for row in state)

def print_state(state):
    for row in state:
        print(row)
    print()

def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

# -------------------- Manhattan Heuristic --------------------

def manhattan(state, goal):
    dist = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                for x in range(3):
                    for y in range(3):
                        if goal[x][y] == val:
                            dist += abs(i - x) + abs(j - y)
    return dist

# -------------------- Neighbors --------------------

def get_neighbors(state):
    x, y = find_blank(state)
    neighbors = []

    for dx, dy, mv in [(-1,0,'U'), (1,0,'D'), (0,-1,'L'), (0,1,'R')]:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append((new_state, mv))

    return neighbors

# -------------------- A* SEARCH --------------------

def A_star(start, goal):
    pq = PriorityQueue()
    visited = set()
    g_cost = {}
    parent = {}

    counter = itertools.count() # Add a counter for tie-breaking

    start_key = to_tuple(start)
    goal_key = to_tuple(goal)

    g_cost[start_key] = 0
    parent[start_key] = (None, None)

    h0 = manhattan(start, goal)
    pq.put((h0, 0, next(counter), start_key, start))   # (f, g, count, key, state)

    while not pq.empty():
        f, g, count, cur_key, cur_state = pq.get() # Get count as well

        if cur_key in visited:
            continue
        visited.add(cur_key)

        if cur_key == goal_key:
            return parent, cur_key

        for nxt_state, mv in get_neighbors(cur_state):
            nxt_key = to_tuple(nxt_state)
            new_g = g + 1

            if nxt_key not in g_cost or new_g < g_cost[nxt_key]:
                g_cost[nxt_key] = new_g
                new_f = new_g + manhattan(nxt_state, goal)
                pq.put((new_f, new_g, next(counter), nxt_key, nxt_state)) # Put count
                parent[nxt_key] = (cur_key, mv)

    return None, None

# -------------------- Reconstruct Path --------------------

def reconstruct(parent, end_key, start):
    path = []
    moves = []

    cur = end_key
    while parent[cur][0] is not None:
        prev, mv = parent[cur]
        path.append(cur)
        moves.append(mv)
        cur = prev

    path.append(to_tuple(start))
    path.reverse()
    moves.reverse()
    return path, moves

# -------------------- RUN --------------------

parent, end_key = A_star(start, goal)

if parent is None:
    print("No solution found.")
else:
    path, moves = reconstruct(parent, end_key, start)
    print("Moves:", "".join(moves))
    print("Total steps:", len(moves))
    print("\nSolution path:")
    for state in path:
        print_state([list(row) for row in state])
```
