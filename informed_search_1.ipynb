## Simple Hill Climbing Algorithm

```python
board=[[1,4,7],
       [2,5,8],
       [3,6,0]]
goal=[[1,2,4],
       [5,0,7],
       [3,6,8]]

#heuristic value=Number of misplace tiles
#h(initial state)=5
```

```python
def heuristic(board, goal):
  misplace=0
  for i in range(len(board)):
    for j in range(len(board[i])):
      if goal[i][j]!=0 and goal[i][j]!=board[i][j]:
        misplace=misplace+1
  return misplace


```

```python
def print_board(board):
  for i in range(len(board)):
    print(board[i])
print_board(board)
```

```python
def blank_space(board):
  for i in range(len(board)):
    for j in range(len(board[i])):
      if board[i][j]==0:
        return i,j


```

```python
x,y=blank_space(board)
print(x,y)
```

```python

h_n=heuristic(board,goal)
print(h_n)
```

```python
def hill_climbing(board,goal,h_n,x,y):
  step=0
  move=[(-1,0),(1,0),(0,-1),(0,1)]
  while True:
    if h_n==0:
      print("Goal Reached")
      return step
    moved=False
    for dx,dy in move:
      n_x=x+dx
      n_y=y+dy
      if (0<=n_x<3 and 0<=n_y<3):
        board[x][y],board[n_x][n_y]= board[n_x][n_y],board[x][y]
        print_board(board)
        h1=heuristic(board,goal)
        print("Heuristic value:", h1)
        if(h1>=h_n):
          board[n_x][n_y],board[x][y]=board[x][y],board[n_x][n_y]
        else:
          h_n=h1
          x,y=n_x,n_y
          moved=True
          step=step+1
          print("Steps: ",step)
          #print_board(board)
          break
    if not moved:
      print("In local optima")
      print_board(board)
      return step




```

```python
step=hill_climbing(board,goal,h_n,x,y)
print("Total Step: ", step)
```

## Steepest Accent Hill Climbing

```python
def steepest_hill_climbing(board, goal, h_n, x, y):
    step = 0
    move = [(-1,0), (1,0), (0,-1), (0,1)]

    while True:
        if h_n == 0:
            print("Goal Reached")
            return step

        moved = False
        x_min, y_min = x, y
        h_min = h_n

        # evaluate all neighbors
        for dx, dy in move:
            n_x = x + dx
            n_y = y + dy

            if 0 <= n_x < 3 and 0 <= n_y < 3:
                # try move
                board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]
                h1 = heuristic(board, goal)
                print_board(board)

                # undo move
                board[x][y], board[n_x][n_y] = board[n_x][n_y], board[x][y]

                # keep best neighbor
                if h1 < h_min:
                    h_min = h1
                    x_min, y_min = n_x, n_y
                    moved = True

        # if no better neighbor exists, stop
        if not moved:
            print("In local optima / plateau")
            return step

        # apply the best move found
        board[x][y], board[x_min][y_min] = board[x_min][y_min], board[x][y]
        x, y = x_min, y_min
        h_n = h_min
        step += 1

        print("Step:", step, "Heuristic:", h_n)
        print_board(board)
```
