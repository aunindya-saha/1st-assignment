## **Greedy Best First Search Algorithm**

```python
edge=int(input("Enter the number of the edges: "))
graph={}
for i in range(edge):
  u, v, w=input("enter the edge and weight: ").split()
  w=int(w)
  if u in graph:
    graph[u].append((v,w))
  else:
    graph[u]=[]
    graph[u].append((v,w))
  if v in graph:
    graph[v].append((u,w))
  else:
    graph[v]=[]
    graph[v].append((u,w))
```

```python
from queue import PriorityQueue

def greedy_bfs(st_node, en_node, graph, heuristic):
    que = PriorityQueue()
    visited = set()  # use local set (faster, avoids global bugs)

    # push start node with its heuristic
    que.put((heuristic.get(st_node, float("inf")), st_node, st_node))

    while not que.empty():
        h_cur, cur_node, path = que.get()
        print(cur_node)

        # goal test
        if cur_node == en_node:
            #print("Path:", path)
            return path

        # expand if not visited
        if cur_node in visited:
            continue
        visited.add(cur_node)

        # push neighbors by heuristic only (greedy)
        for nei in graph.get(cur_node, []):
            nxt = nei[0]              # neighbor city name
            if nxt not in visited:
                h_nxt = heuristic.get(nxt, float("inf"))
                que.put((h_nxt, nxt, path + " " + nxt))

    print("No path found.")
    return None


# Example usage:
st_node = input("enter the first node: ").strip()
en_node = input("enter the goal node: ").strip()

greedy_bfs(st_node, en_node, graph, heuristic)
```

## A* **Search**

```python
from queue import PriorityQueue

def A_star(st_node, en_node, graph, heu):
    que = PriorityQueue()
    visited = set()

    # (f, g, node, path)
    que.put((heuristic.get(st_node, float("inf")), 0, st_node, st_node))

    while not que.empty():
        f, g, cur_node, path = que.get()
        print(cur_node)

        # Goal test
        if cur_node == en_node:
            print("Path:", path)
            print("Total cost:", g)
            return path

        if cur_node in visited:
            continue
        visited.add(cur_node)

        for nei in graph.get(cur_node, []):
          nxt=nei[0]
          cost=nei[1]
          if nxt not in visited:
                new_g = g + cost
                new_f = new_g + heu.get(nxt, float("inf"))
                que.put((new_f, new_g, nxt, path + " " + nxt))

    print("No path found.")
    return None
```

```python
st_node = input("enter the first node: ").strip()
en_node = input("enter the goal node: ").strip()

A_star(st_node, en_node, graph, heuristic)
```
